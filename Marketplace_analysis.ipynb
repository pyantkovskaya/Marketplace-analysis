## Задача 1. Оценить месячный retention в оформление заказа с помощью когортного анализа.

* Исследовать датасет и определить, какой вид заказа будет учитываться в retention

* Построить месячный retention

* Проанализировать, чему равен медианный retention 1-го месяца (т.е. взять retention 1-ого месяца для всех когорт и посчитать по этим числам медиану)

* Найти когорту с самым высоким retention на 3-й месяц.

```python
data = olist_orders_dataset[['customer_id', 'order_purchase_timestamp', 'order_status']]\
        .query('order_status == "delivered"')\
        .merge(olist_customers_dataset, on='customer_id')
```

```python
data.drop(['customer_zip_code_prefix', 'customer_city', 'customer_state', 'order_status', 'customer_id'], axis=1, inplace=True)
```

```python
data['order_month'] = data['order_purchase_timestamp'].dt.to_period('M')
data['order_day'] = data['order_purchase_timestamp'].dt.to_period('D')
```

```python
#Присаиваем когорты на основе даты первого действия пользователя
data['cohort'] = data.groupby('customer_unique_id')['order_purchase_timestamp'].transform('min').dt.to_period('M')
data['cohort_day'] = data.groupby('customer_unique_id')['order_purchase_timestamp'].transform('min').dt.to_period('D')
```

```python
# Расчитываем номер периода в днях (месячные периоды 30 дней)
data['period_number_month'] = np.floor((data.order_day - data.cohort_day).apply(attrgetter('n')) / 30)
```

```python
# Расчитываем число покупателей по когортам
df_cohort = data.groupby(['cohort', 'period_number_month']).agg(n_customers=('customer_unique_id', 'nunique')).reset_index()
```

```python
cohort_pivot = df_cohort.pivot_table(index='cohort', columns='period_number_month', values='n_customers')
```

```python
cohort_size = cohort_pivot.iloc[:, 0]
cohort_size
```

```python
retention_matrix = cohort_pivot.divide(cohort_size, axis=0)
```

```python
# Строим тепловую карту когортного анализа
with sns.axes_style("white"):
    fig, ax = plt.subplots(1, 2, figsize=(22, 12), sharey=True, gridspec_kw={'width_ratios': [1, 11]})

    # Удаляем первый месяц (где retention 100%)
    retention_matrix_filtered = retention_matrix.iloc[:, 1:]
    
    # Настраиваем цветовую карту для малых значений
    custom_cmap = sns.color_palette("YlGnBu", as_cmap=True)
    
    # Тепловая карта для коэффициентов удержания
    sns.heatmap(retention_matrix_filtered,
                mask=retention_matrix_filtered.isnull(),
                annot=True,
                fmt='.2%', 
                cmap=custom_cmap,
                ax=ax[1],
                annot_kws={"size": 11.5},
                vmin=0,  # Минимальное значение
                vmax=0.01,  # Максимальное значение (1%)
                center=0.005)  # Центр цветовой шкалы
    
    ax[1].set_title('Monthly Retention (Excluding First Month)', fontsize=20)
    ax[1].set(xlabel='№ периода', ylabel='Когорта')

    # Тепловая карта для размеров когорт (тоже убираем первый месяц)
    cohort_size_df = pd.DataFrame(cohort_size).rename(columns={0: 'cohort_size'})
    white_cmap = mcolors.ListedColormap(['white'])
    sns.heatmap(cohort_size_df,
                annot=True,
                cbar=False,
                fmt='g',
                cmap=white_cmap,
                ax=ax[0])

    fig.tight_layout()
    plt.show()
```

```python
# Медианный retention 1-го месяца 
median_retention_1m = retention_matrix.iloc[:, 1].median()
print(f"Медианный retention 1-го месяца: {median_retention_1m:.2}")
```

```python
# Когорта с самым высоким retention на 3-й месяц
max_retention_3m_cohort = retention_matrix.iloc[:, 3].idxmax()
print(f"Когорта с самым высоким retention на 3-й месяц: {max_retention_3m_cohort}")
```

## Задача 3. Определить 5 основных метрик, на которых продакт может сконцентрироваться, чтобы максимизировать прибыль компании.

В первую очередь необходимо:

Определить 5 основных метрик, на которых продакт может сконцентрироваться, чтобы максимизировать прибыль компании.

* Первая метрика должна отражать рост объёма продаж маркетплейса.
* Вторая — показывать объем аудитории, которой продукт доставляет ценность.
* Третья — отражать заинтересованность новых клиентов в продакте.
* Четвёртая — отражать вовлеченность клиента в продолжение использования продукта.
* Пятая — отражать денежное выражение вовлеченности клиента.

Визуализируйте первую, вторую, четвёртую и пятую метрики. Используйте месячную гранулярность и окно в 1 месяц, если это нужно.

### Пять ключевых метрик для максимизации прибыли:
    
1. **GMV (Gross Merchandise Value)** – общий объем продаж на маркетплейсе. Можно сравнить GMV по месяцам.

    *GMV = Цена × Количество проданных товаров*


2. **MAU (Monthly Active Users)** – количество активных пользователей в месяц.

3. **CR (Conversion Rate)** – конверсия новых пользователей в покупателей.

    *CR = Кол-во новых пользователей, совершивших покупку / Общее кол-во новых пользователей × 100%*
    

4. **RPR (Repeat Purchase Rate)** – доля клиентов, совершивших повторную покупку.

    *RPR = Количество клиентов, совершивших повторную покупку / Общее количество клиентов × 100%*
    

5. **ARPPU (Average Revenue Per Paying User)** — средний доход с клиента

    *ARPU = GMV / Количество уникальных покупателей*



```python
# GMV 
# Для расчета продаж возьмем только доставленные заказы
delievered_orders = olist_orders_dataset.query('order_status == "delivered"')
```

```python
# Сгруппируем данные по месяцам создания заказа и доставки (если оплата по факту доставки и важен момент поступления денег)

orders_with_cost['purchase_month'] = orders_with_cost['order_purchase_timestamp'].dt.to_period("M")
orders_with_cost['delievery_month'] = orders_with_cost['order_delivered_customer_date'].dt.to_period("M")
GMV  = orders_with_cost.groupby('purchase_month').order_cost.sum()
GMVd  = orders_with_cost.groupby('delievery_month').order_cost.sum()
```
